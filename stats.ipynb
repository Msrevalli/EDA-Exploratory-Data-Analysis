# Central Limit Theorm Proof

```python
sample_mean = []
for i in range(1,35):
    sample = np.random.choice(population,size)
    sample_mean.append(np.mean(sample))
```

```python
sns.kdeplot(sample_mean)
```

```python
sample_mean = []
for i in range(1,100):
    sample = np.random.choice(population,size)
    sample_mean.append(np.mean(sample))

sns.kdeplot(sample_mean)
```

```python
sample_mean = []
for i in range(1,1000):
    sample = np.random.choice(population,size)
    sample_mean.append(np.mean(sample))

sns.kdeplot(sample_mean)
```

```python
sample_mean = []
for i in range(1,1000):
    sample = np.random.choice(population,size)
    sample_mean.append(np.mean(sample))

sns.histplot(sample_mean, kde=True)
```
